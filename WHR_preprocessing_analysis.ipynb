```python
# Only for 2018 as there was a single NaN in gov_trust column which would throw an error while calculating 'Dystopia Residual' due to dtype mismatch. 
df['govt_trust'].fillna(0.0, inplace=True)
df['govt_trust'] = df['govt_trust'].astype(float)
```

```python
#2018, 2019
df['dystopia_residual'] = df['happiness_score'] - (df['economy'] + df['family'] + df['life_expectancy'] + df['freedom'] + df['govt_trust'] + df['generosity'])
```

```python
#creating a combined dataset
updated_2015 = pd.read_csv('updated_2015.csv')
updated_2016 = pd.read_csv('updated_2016.csv')
updated_2017 = pd.read_csv('updated_2017.csv')
updated_2018 = pd.read_csv('updated_2018.csv')
updated_2019 = pd.read_csv('updated_2019.csv')

updated_2015['year'] = 2015
updated_2016['year'] = 2016
updated_2017['year'] = 2017
updated_2018['year'] = 2018
updated_2019['year'] = 2019

combined_file = pd.concat([updated_2015, updated_2016, updated_2017, updated_2018, updated_2019], ignore_index = True)
combined_file
```

```python
combined_file = combined_file[['year', 'country', 'region1', 'continent', 'happiness_rank', 'happiness_score', 'economy', 'family', 'life_expectancy', 
                              'freedom', 'govt_trust', 'generosity', 'dystopia_residual']]

combined_file.to_csv('WHR_combined_dataset.csv', index=False)
```
